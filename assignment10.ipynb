# Assignment

1. Create a border around any chosen image, make sure the border is 10% of the image shape. (border can be any color)
2. At the centre of the image create a yellow square of 20% minimum of the image shape.
3. Create a gray image from Q1 image and draw a square of white (10%) at four corners of the image.
4. Make sure you save the two resulting images as PNG.

```python
# 1. Create a border around any chosen image, make sure the border is 10% of the image shape. (border can be any color)
# 2. At the centre of the image create a yellow square of 20% minimum of the image shape.

import cv2 as cv
import sys

try:
    path = r'../essential/assets/myimage2.jpg' # r notation for raw string
    img = cv.imread(path, cv.IMREAD_COLOR)

    if img is None:
        sys.exit("Could not read the image.")
    
    # Check image shape and size
    imgShape = img.shape
    imgSize = img.size
    
    print('Shape:', imgShape)
    print('Size:', imgSize)
    
    # Define border size
    bsize_row = int(imgShape[0]*0.1)
    bsize_col = int(imgShape[1]*0.1)
    
    # Define the centre of image
    middle_row = int(imgShape[0]/2)
    middle_col = int(imgShape[1]/2)
    
    print(f'Border row size: {bsize_row}\nBorder column size: {bsize_col}')
    
    # To make the border equal in size
    if bsize_row < bsize_col: # Just change < or > if needed
        bsize_col = bsize_row
    else:
        bsize_row = bsize_col
    
    # Display before edit
    cv.imshow("Display window", img)
    k = cv.waitKey(0)

    # Make a copy of the original image
    imgWithBorder = img.copy()
    
    # Color the border
    imgWithBorder[:bsize_row,:] = [244, 133, 66] # Top border
    imgWithBorder[:,-bsize_col:] = [88, 157, 15] # Right border
    imgWithBorder[-bsize_row:,:] = [0, 230, 230] # Bottom border
    imgWithBorder[bsize_row:,:bsize_col] = [55, 68, 219] # Left border
    
    # Middle square
    imgWithBorder[middle_row - bsize_row : middle_row + bsize_row, 
                  middle_col - bsize_col : middle_col + bsize_col] = [0,255,255] 
    
    
    # Display after edit
    cv.imshow('Image with border and square centre', imgWithBorder)
    cv.waitKey(0)
    
    # Save image
    cv.imwrite("myimage_bordered.png", imgWithBorder)
    
except Exception as e:
    print(e)
    
finally:
    cv.destroyAllWindows()
```

```python
# 3. Create a gray image from Question 1 image and draw a square of white (10%) at four corners of the image.

import cv2 as cv
import sys

try:
    path = r'../essential/assets/myimage2.jpg' # r notation for raw string
    img = cv.imread(path, cv.IMREAD_GRAYSCALE) # Read as grayscale
    
    # Check image shape and size
    imgShape = img.shape
    imgSize = img.size
    
    print('Shape:', imgShape)
    print('Size:', imgSize)
    
    # Define border size
    bsize_row = int(imgShape[0]*0.1)
    bsize_col = int(imgShape[1]*0.1)

    print(f'Border row size: {bsize_row}\nBorder column size: {bsize_col}')
    
    # To make the border equal in size
    if bsize_row < bsize_col: # Just change < or > if needed
        bsize_col = bsize_row
    else:
        bsize_row = bsize_col
    
    # Display before edit
    cv.imshow("Display window", img)
    k = cv.waitKey(0)
    

    # Make a copy of the original image
    imgWithBorder = img.copy()
    
    # Color the corner
    imgWithBorder[:bsize_row,-bsize_col:] = 255 # Top right
    imgWithBorder[:bsize_row,:bsize_col] = 255 # Top left
    imgWithBorder[-bsize_row:,-bsize_col:] = 255 # Bottom right
    imgWithBorder[-bsize_row:,:bsize_col] = 255 # Bottom left
    
    # Display after edit
    cv.imshow('Grayed and white cornered image', imgWithBorder)
    cv.waitKey(0)
    
    # Save image
    cv.imwrite("myimage_grayed.png", imgWithBorder)
    
except Exception as e:
    print(e)
    
finally:
    cv.destroyAllWindows()
```
