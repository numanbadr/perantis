1. Create a 2D data of shape 10x5x3 with values equals zero
2. Create another array with values 255 and shape 10x5x3
3. Join the 2 array horizontally
4. Create an array of random unsigned integer 8 bits with shape (50x50x3)
5. Convert top part to red (255,0,0) and bottom part to green (0,255,0)
6. Seed value is 11
7. Display the data

```python
import numpy as np
import matplotlib.pyplot as plt

# 1 & 2 Create zero amd 255 array

zero_array = np.zeros((10,5,3), dtype='uint8')
array_255 = np.full((10,5,3),255)

# Alternative ways to create array_255
# array_255 = np.where(zero_array==0,255, a)
# array_255 = np.ones((10,5,3))*255

# 3. Join the array 
h_joined = np.hstack((zero_array, array_255))
v_joined = np.vstack((zero_array, array_255))


# Just to display the difference when we do hstack and vstack
plt.subplot(1,2,1)
plt.imshow(h_joined)
plt.title('Horizontally joined')

plt.subplot(1,2,2)
plt.imshow(v_joined)
plt.title('Vertically joined')
plt.show()
```

```python
# 4. Create an array of random unsigned integer 8 bits with shape (50x50x3)
np.random.seed(11)
b = np.random.randint(0, 256, (50, 50, 3))

plt.figure(figsize=(10,20))
plt.imshow(b)
plt.title('ORIGINAL ARRAY')
plt.show()
```

```python
# 5. Convert top part to red (255,0,0) and bottom part to green (0,255,0)

# Identify top and bottom row
top = b[:1,:,:]
bottom = b[-1:,:,:]

# Create an array of row for red and green values
red = np.full((1,50,3),[255,0,0])
green = np.full((1,50,3),[0,255,0])

# Replace the top and bottom row with red and green
b_new = np.copy(b)
b_new = np.where(b==top, red, b_new)
b_new = np.where(b==bottom, green, b_new)

# Display the result
print(top.shape)
plt.figure(figsize=(10,20))
plt.imshow(b_new)
plt.title('RED TOP GREEN BOTTOM ARRAY')
plt.show()
```

```python
# EXTRA: Check 2 rows top and bottom to compare BEFORE AND AFTER
# The colour of 2nd top and 2nd bottom rows remain the same. 
# Therefore, objective to change only top and bottom row is achieved.

plt.figure(figsize=(20,7))

plt.subplot(4,1,1)
plt.imshow(b_new[:2,:,:])
plt.title('TOP ROW AFTER')

plt.subplot(4,1,2)
plt.imshow(b[:2,:,:])
plt.title('TOP ROW BEFORE')

plt.show()
print()
plt.figure(figsize=(20,7))

plt.subplot(4,1,3)
plt.imshow(b[-2:,:,:])
plt.title('BOTTOM ROW BEFORE')

plt.subplot(4,1,4)
plt.imshow(b_new[-2:,:,:])
plt.title('BOTTOM ROW AFTER')

plt.show()
```
